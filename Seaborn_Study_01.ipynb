```python
# 시각화 라이브러리
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = 'darkgrid')
```

```python
# 폰트 조정
import matplotlib.font_manager as fm
font_list = [font.name for font in fm.fontManager.ttflist]
font_list # 변견가능한 폰트 리스트 출력
plt.rcParams['font.family'] = 'Malgun Gothic' # 원하는 폰트로 변경
```

```python
# 주피터 노트북에서 사용시 내부에 그림이 바로 그려지도록 하는 명령어
# %matplotlib inline 
```

```python
# 데이터셋 (Seaborn 기본적으로 제공하는 데이터)
tips = sns.load_dataset('tips')
tips
```

산점도 그래프 - relplot

```python
# 산점도 그래프 - relplot
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

relplot 옵션

```python
# hue - 그려진 그래프 내에서 서로 다른 색으로 마커 표현
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size')
```

```python
# style - 그려진 그래프 내에서 서로 다른 기호로 마커 표현
sns.relplot(x = 'total_bill', y = 'tip', data = tips, style = 'smoker')
```

```python
# style과 hue 함께 사용 가능
sns.relplot(x = 'total_bill', y = 'tip', data = tips, style = 'smoker', hue = 'size')
```

```python
# palette - 마커 색상 변경
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', palette = 'Set2')

# palette 옵션은 seaborn 사이트에서 확인 가능
```

```python
# size - 마커 크기 변경 (size와 hue 함께 사용 불가)
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size')
```

```python
# relplot( ).set(xlabel = ' ' , ylabel = ' ' ) - X축/Y축 이름 변경
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size').set(xlabel = '총합', ylabel = '팁')
```

```python
import pandas as pd
import numpy as np
```

```python
# 임의의 데이터 프레임 만들기
df = pd.DataFrame(dict(time = np.arange(500), value = np.random.randn(500).cumsum()))
df
```

선 그래프 - relplot ( kind = 'line' )

```python
# 선 그래프 - sns.relplot(kind = 'line')
sns.relplot(x = 'time', y = 'value', data = df, kind = 'line')
```

```python
# 다른 데이터 불러오기
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri)

# 가운데 실선 : 평균값
# 실선 주변 범위값 : 표준편차 범위
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = None)

# ci = None 으로 평균값만 표현
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = 'sd')

# ci = 'sd' 분산값 범위 표현
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, hue = 'region', style = 'event', ci = None)
```
