```python
# 데이터셋 (Seaborn 기본적으로 제공하는 데이터)
tips = sns.load_dataset('tips')
tips
```

relplot
- 관계형 데이터 시각화 시 주로 사용  
   ex) 시간에 따른 데이터 변화, 인원수에 따른 데이터 변화 등등

```python
# 산점도 그래프(관계형 데이터) - relplot
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

relplot 옵션

```python
# hue - 그려진 그래프 내에서 서로 다른 색으로 마커 표현
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size')
```

```python
# style - 그려진 그래프 내에서 서로 다른 기호로 마커 표현
sns.relplot(x = 'total_bill', y = 'tip', data = tips, style = 'smoker')
```

```python
# style과 hue 함께 사용 가능
sns.relplot(x = 'total_bill', y = 'tip', data = tips, style = 'smoker', hue = 'size')
```

```python
# palette - 마커 색상 변경
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', palette = 'Set2')

# palette 옵션은 seaborn 사이트에서 확인 가능
```

```python
# size - 마커 크기 변경 (size와 hue 함께 사용 불가)
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size')
```

```python
# relplot( ).set(xlabel = ' ' , ylabel = ' ' ) - X축/Y축 이름 변경
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size').set(xlabel = '총합', ylabel = '팁')
```

```python
import pandas as pd
import numpy as np
```

```python
# 임의의 데이터 프레임 만들기
df = pd.DataFrame(dict(time = np.arange(500), value = np.random.randn(500).cumsum()))
df
```

선 그래프 - relplot ( kind = 'line' )

```python
# 선 그래프 - sns.relplot(kind = 'line')
sns.relplot(x = 'time', y = 'value', data = df, kind = 'line')
```

```python
# 다른 데이터 불러오기
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri)

# 가운데 실선 : 평균값
# 실선 주변 범위값 : 표준편차 범위
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = None)

# ci = None 으로 평균값만 표현
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = 'sd')

# ci = 'sd' 분산값 범위 표현
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, hue = 'region', style = 'event', ci = None)
```

catplot
- 범주형 데이터 시각화 시 주로 사용  
   ex) 성별 데이터, 혈액형 데이터 등등

```python
# 산점도 그래프(범주형 데이터) - catplot
sns.catplot(x = 'day', y = 'total_bill', data = tips) # 요일별 요금정산 금액
                                                      # defult 값 : ( kind = 'strip' )
```

catplot 옵션

```python
# catplot ( kind = 'swam' ) - 마커를 겹쳐지지 않게 표기
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm') # 요일별 요금정산 금액
```

```python
# catplot ( hue = 'sex', kind = 'swam' ) - 성별 데이터 구분해서 표기
sns.catplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', kind = 'swarm') # 요일별 요금정산 금액
```

```python
# catplot ( kind = 'violin' ) - 좌 우 데이터 비교 가능
sns.catplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', inner = 'stick', split = True, kind = 'violin') # 요일별 요금정산 금액
# split 옵션으로 합치기 가능
# inner 옵션으로 실제 데이터를 가로선으로 표기
```

```python
# catplot ( kind = 'box' ) - 0% 25% 50% 75% 100% 이상치 표기
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'box') # 요일별 요금정산 금액
```

```python
# catplot ( kind = 'boxen' ) - box에 비해 분포를 보다 정확하게 시각화하도록 설계됨
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'boxen') # 요일별 요금정산 금액
```

```python
# catplot ( kind = 'bar' ) - 막대그래프 (중앙 실선은 편차를 나타냄)
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'bar') # 요일별 요금정산 금액
```

```python
# catplot ( kind = 'violin' ) - 좌 우 데이터 비교 가능
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'point') # 요일별 요금정산 금액
```
